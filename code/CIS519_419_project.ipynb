```python
models = ["svm", "scanvi", "XGBoost", "LogisticRegression", "RandomForest"]
datasets = ["Xin", "Segerstolpe", "Baron Human", "ALL"]

values = np.zeros((len(datasets), len(models)))

funs = [tools.run_svm, tools.run_scanvi, tools.run_XGBoost, tools.run_logistic_regression, tools.run_random_forest]
paths = ['Xin.h5ad', 'Segerstolpe.h5ad', 'Baron Human.h5ad', 'ALL.h5ad']

for i in range(len(datasets)):
    adata = tools.preproccess(paths[i])
    for j in range(len(models)):
        if datasets[i] == 'ALL':
            values[i,j] = round(funs[j](adata, labels_key='Ctype_Final')[0], 3)
        else:
            values[i,j] = round(funs[j](adata)[0], 3)

tools.draw_heatmap(models, datasets, values)
```
